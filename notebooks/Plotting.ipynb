```python
# stdlib imports
import glob

# third party imports
import matplotlib.pyplot as plt

# local imports
from amptools.io.read import read_data
from amptools.plot import plot_arias, plot_durations, plot_moveout
```

# Plotting Tools

```python
# read in data
event_dir = '../tests/data/cwb'
streams = []
for filename in glob.glob(event_dir + '/*'):
    streams += [read_data(filename)]
```

## Variations of Arias Intensity Plots

#### Full default plot:

```python
axes = plot_arias(streams[3])
```

#### Combining multiple plots:

```python
fig, axs = plt.subplots(len(streams), 3, figsize=(15,10))
axs = axs.flatten()
idx = 0
for stream in streams:
    axs = plot_arias(stream, axes=axs, axis_index=idx, minfontsize=15,
                        show_maximum=False, title="18km NNE of Hualian, Taiwan")
    idx += 3
```

## Variations of Duration Plots

#### Full default plot:

```python
durations = [(0.05, 0.75),
                 (0.2, 0.8),
                 (0.05, .95)]
axes = plot_durations(streams[3], durations)
```

#### Combining multiple plots:

```python
fig, axs = plt.subplots(len(streams), 3, figsize=(15,10))
axs = axs.flatten()
idx = 0
for stream in streams:
    axs = plot_durations(stream, durations, axes=axs, axis_index=idx, minfontsize=15, title="18km NNE of Hualian, Taiwan")
    idx += 3
```

## Moveout plots

#### Full default plot:

```python
epicenter_lat = 24.14
epicenter_lon = 121.69
fig, ax = plot_moveout(streams, epicenter_lat, epicenter_lon, 'BN1')
```

#### Moveout with scaling and specified colormap:
The data is automatically normalized if a scaling factor is specified.

```python
fig, ax = plot_moveout(streams, epicenter_lat, epicenter_lon, 'BN1', cmap='nipy_spectral_r', figsize=(15, 10),
        minfontsize=16, normalize=True, scale=10)
```
